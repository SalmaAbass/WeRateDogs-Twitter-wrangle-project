## Analyzing, and Visualizing WeRateDogs

## Introduction:
This analysis is of tweets from the WeRateDogs twitter account. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. It was started in 2015 by college student Matt Nelson


## Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('twitter_archive_master.csv')
df['timestamp']=pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
```

### Question 1
What is the most popular dog breed in our data set?

```python
df['dog_type'].value_counts().head(10).plot(kind='barh')
```

Golden Retrievers are the most popular breed on the WeRateDogs Twitter feed, followed by Pembrokes and Labrador Retrievers.
This  could be because they are commonly owned or they could be the favorites of the owner of @dog_rates, so the ones they choose to rate the most.


### Question 2
*What is the top 10 rating dogs breed on average.*

```python
df.groupby('dog_type').rating.mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.xlabel('average_rating')
plt.ylabel('Top 10 dogs bread')
plt.title('Top 10 rating dogs bread')
plt.xticks(rotation=70)

plt.savefig('top_10_rating_dogs_bread.png')
```

### Question 3
*What is the lowest rating dog breed?* 
Japanese_spaniel has the lowest rating equal to 5.

```python
df.groupby('dog_type').rating.mean().sort_values(ascending=False).tail(1)
```

### Question 4
What is distribution of retweet and favorite for each dogs stage?
Here we can see that pupper and floofer dogs on average have higher retweet count and favorite count.

```python
width=0.4
retweet_mean = df.groupby('dog_stage').retweet_count.mean()
favorite_mean = df.groupby('dog_stage').favorite_count.mean()
x_axis = np.arange(4)
plt.bar(x_axis,retweet_mean,width,alpha=.7,label='Average_ retweet_count')
plt.bar(x_axis+width,favorite_mean,width,alpha=.7,label='Average_favorite_count')
plt.xticks(x_axis+width/2,['doggo','puppo','pupper','floofer'])
plt.legend(bbox_to_anchor=(0.25,1),
           loc='lower left')
plt.savefig('distribution_of_retweet_and_favorite_for_each_dogs_stage.png')
plt.show()
```

### Question 5
Does higher ratings leads to higher retweets or higher favorite count? 
There is no correlation between rating and retweets or favorites. 

```python
df.plot(x='retweet_count', y='rating', kind='scatter')
plt.title('Retweet Counts by Ratings Scatter Plot')
plt.xlabel('Retweet Counts')
plt.ylabel('Ratings')
plt.savefig('retweet_vs_rating.png')
plt.ylim([0,20])
```

```python
df.retweet_count.corr(df.rating)
```

```python
df.plot(x='favorite_count', y='rating', kind='scatter')
plt.title('favorite_count by Ratings Scatter Plot')
plt.xlabel('favorite_count')
plt.ylabel('Ratings')
plt.savefig('favorite_count vs_rating.png')
plt.ylim([0,20])
```

```python
df.favorite_count.corr(df.rating)
```

### Question 6
